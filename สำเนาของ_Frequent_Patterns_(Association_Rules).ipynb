````python
data['InvoiceDate'][0].split()[0] #split  ใช้เเยกเวลา ออกจากวัน
````

````python
date = []
for i in range(len(data['InvoiceDate'])):
    x = data['InvoiceDate'][i].split()[0]
    date.append(x)

L = set(date) #ลูปข้อมูลทั้งหมด
````

````python
date[0].split('/')
````

###เเยกวัน เดือน ปีออกจากกัน

````python
day =[]
month =[]
year =[]
for i in range(len(data['InvoiceDate'])):
  x = date[i].split('/')[0]
  y = date[i].split('/')[1]
  z = date[i].split('/')[2]
  month.append(x)
  day.append(y)
  year.append(z)
````

ซึ่งจะถือว่ารายการ 'วันที่' มีสตริงที่มีรูปแบบเช่น 'dd/mm/yyyy' (วัน/เดือน/ปี) ปรับโค้ดให้เหมาะสมหากรูปแบบวันที่ของคุณแตกต่างออกไป

หลังจากรันโค้ดนี้แล้ว คุณควรมีรายการ 'วัน', 'เดือน' และ 'ปี' แยกกัน ซึ่งมีค่าที่สอดคล้องกันซึ่งดึงมาจากรายการ 'วันที่'
